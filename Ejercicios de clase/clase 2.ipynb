2. NÚMERO DE ACTUALS Y FORECAST

```python
import pandas as pd

# Carga el archivo CSV
file_path = 'datos_ejercicio_ventas.csv'  # Cambia esta ruta por la ubicación real de tu archivo
data = pd.read_csv(file_path)

# Cuenta las ocurrencias de 'actual' y 'AI_forecast' en la columna 'SCENARIO'
scenario_counts = data['SCENARIO'].value_counts()

# Muestra los resultados
actual_count = scenario_counts.get('actual', 0)
ai_forecast_count = scenario_counts.get('AI_forecast', 0)

print(f"Número de 'actual': {actual_count}")
print(f"Número de 'AI_forecast': {ai_forecast_count}")
```

5. HISTÓRICO DE ACTUALS Y FORECAST

```python
# Rango de fechas para los forecasts (AI_forecast)
forecast_df = data[data['SCENARIO'] == 'AI_forecast']  # Cambié df a data

# Obtener el año y mes mínimo y máximo para las predicciones (forecasts)
min_forecast_year = forecast_df['YEAR'].min()
max_forecast_year = forecast_df['YEAR'].max()

min_forecast_month = forecast_df[forecast_df['YEAR'] == min_forecast_year]['MONTH'].min()
max_forecast_month = forecast_df[forecast_df['YEAR'] == max_forecast_year]['MONTH'].max()

(min_forecast_year, min_forecast_month), (max_forecast_year, max_forecast_month)
```

```python
# Rango de fechas para los forecasts (AI_forecast)
actual_df = data[data['SCENARIO'] == 'actual']  # Cambié df a data

# Obtener el año y mes mínimo y máximo para las predicciones (forecasts)
min_actual_year = actual_df['YEAR'].min()
max_actual_year = actual_df['YEAR'].max()

min_actual_month = actual_df[actual_df['YEAR'] == min_actual_year]['MONTH'].min()
max_actual_month = actual_df[actual_df['YEAR'] == max_actual_year]['MONTH'].max()

(min_actual_year, min_actual_month), (max_actual_year, max_actual_month)
```

6. CUÁNTOS FORECAST TIENE (EL AÑO CUENTA)

```python
# Group by 'YEAR' and 'FORECAST' and then count distinct values of 'FORECAST' per year
distinct_forecasts_per_year = data.groupby('YEAR')['FORECAST'].nunique().reset_index()
distinct_forecasts_per_year.columns = ['Year', 'Distinct Forecasts']
distinct_forecasts_per_year
```
